```python
import numpy as np
from matplotlib import pyplot as plt 
np.random.seed(69)
```

```python
s = np.random.uniform(-np.pi,np.pi,40)
noise = np.random.normal(0,0.2,40)
Y = np.sin(s)+noise 
dataset = np.c_[s,Y]
plt.scatter(dataset[:,0],dataset[:,1])
f = lambda x:np.sin(x)
x = np.linspace(-np.pi,np.pi,40)
plt.plot(x,f(x),c='g') # target function
```

```python
index1= np.random.choice(dataset.shape[0],size=10,replace=False)
index2= np.random.choice(dataset.shape[0],size=10,replace=False)
index3= np.random.choice(dataset.shape[0],size=10,replace=False)
index4= np.random.choice(dataset.shape[0],size=10,replace=False)
data1x = dataset[index1]
data2x = dataset[index2]
data3x = dataset[index3]
data4x = dataset[index4]
plt.scatter(data1x[:,0],data1x[:,1])
plt.scatter(data2x[:,0],data2x[:,1])
plt.scatter(data3x[:,0],data3x[:,1])
plt.scatter(data4x[:,0],data4x[:,1])
```

# Model with lower complexity

```python
#  model complexity is higher
# H1(x) = w1
h1 = lambda x: 0*x+0.9 #hypothesis 1
h2 = lambda x: 0*x+0.5 #hypothesis 2
h3 = lambda x: 0*x-0.4 #hypothesis 3
h4 = lambda x: 0*x-0.8 #hypothesis 4

plt.plot(x,h1(x),c='b')
plt.plot(x,h2(x),c='r')
plt.plot(x,h3(x),c='g')
plt.plot(x,h4(x),c='y')
plt.plot(x,f(x),c='black')
plt.scatter(s,Y)
g = (h1(x)+h2(x)+h3(x)+h4(x))/4
```

```python
plt.plot(x,h1(x),label='hypothesis 1')
plt.plot(x,h2(x),label='hypothesis 2')
plt.plot(x,h3(x),label='hypothesis 3')
plt.plot(x,h4(x),label='hypothesis 4')
plt.plot(x,g,c='black',label='mean hypothesis')
plt.legend()
plt.plot(x,f(x),c='cyan')
```

```python
plt.plot(x,f(x),label='target hypothesis')
plt.plot(x,g,c='black',label='mean hypothesis')
plt.scatter(s,Y)
plt.legend()
```

```python
var = ((h1(x)-g)**2 + (h2(x)-g)**2 + (h3(x)-g)**2 + (h4(x)-g)**2).mean()
bias = ((g-f(x))**2).mean()
'bias: ',bias,' variance: ',var
```

```python
bias+var
```

# Model with high model complexity

```python
# model complexity lower 

h_1 = lambda x : x
h_2 = lambda x : -(1.5/3)*x +1.5
h_3 = lambda x : -(1.5/3)*x -1.5
h_4 = lambda x : 0.2*x 


plt.plot(x,h_1(x),label='hypothesis 1')
plt.plot(x,h_2(x),label='hypothesis 2')
plt.plot(x,h_3(x),label='hypothesis 3')
plt.plot(x,h_4(x),label='hypothesis 4')
plt.scatter(s,Y)
plt.legend()

g_ = (h_1(x)+h_2(x)+h_3(x)+h_4(x))/4
# plt.plot(x,g_,c='black') # the mean hypothesis
```

```python
plt.plot(x,f(x),c='black',label='target fn f(x)')
plt.plot(x,g_,label='expected h(x)')
plt.scatter(s,Y)
plt.legend()
```

```python
bias2 = ((g_-f(x))**2).mean(axis=0)
'Bias: ',bias2
```

```python
var2 = ((h_1(x)-g_)**2 + (h_2(x)-g_)**2 + (h_3(x)-g_)**2 + (h_4(x)-g_)**2).mean(axis=0)
'variance: ',var2
```

```python

bias2+var2
```
